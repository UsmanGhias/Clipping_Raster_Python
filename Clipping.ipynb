```python
from osgeo impport gdal, osr
```

```python
# Input file path for the source geospatial data (e.g., a GeoTIFF file)
fn_in = r"images/input.tif"

# Output file path for the processed or cropped geospatial data
fn_extent = 'images/output.tif'

# UTM (Universal Transverse Mercator) bounds in the form of (min_x, min_y, max_x, max_y)
bounds_utm11 = (641000., 4851000., 649000., 4859000.)
```

```python
# Create a Spatial Reference object for UTM Zone 11 with the NAD83 datum
utm11_srs = osr.SpatialReference()

# Set the Well-Known GeogCS (Geographic Coordinate System) to NAD83
utm11_srs.SetWellKnownGeogCS("NAD83")

# Set the UTM (Universal Transverse Mercator) projection with Zone 11 and the Northern Hemisphere (1)
utm11_srs.SetUTM(11, 1)
```

```python
# Use GDAL's Warp function to create a new geospatial dataset with specified parameters
gdal.Warp(
    fn_extent,           # Output file path for the warped dataset
    fn_in,               # Input file path of the source dataset
    outputBounds=bounds_utm11,  # Define the output bounds using the provided UTM11 coordinates
    outputBoundsSRS=utm11_srs  # Specify the Spatial Reference System (SRS) for the output bounds
)
```

# $$ Clip Raster to Polygon $$

```python
from osgeo import gdal

# Import the GDAL library, which is used for working with geospatial data

fn_in = r"path/to/input.tif"
# Define the variable fn_in with the path to an input geospatial dataset (a raster file in TIFF format).

fn_clip = 'path/to/output.tif'
# Define the variable fn_clip with the path to the output geospatial dataset (a raster file in TIFF format). This is where the clipped result will be saved.

fn_poly = r"path/to/polygon.gpkg"
# Define the variable fn_poly with the path to a polygon dataset (a vector file in GeoPackage format). This polygon will be used for clipping the input raster data.

# Please note that the actual file paths should be provided in place of "path/to/input.tif" and others.
```

```python
gdal.Warp(fn_clip, fn_in, cutlineDSName=fn_poly, cutlineLayer='polygon', cropToCutline=True)
```
